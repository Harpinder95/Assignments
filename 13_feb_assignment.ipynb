### Q1. Explain why we have to use the Exception class while creating a Custom Exception.

#### While we creates the custom exceptions then, it is mandatory to inherit the Exception class because  user defined classes cannot be directly derived from the baseException class, to derive user defied class, we need to use Exception class. The example of above said is given below:  

```python
import logging
logging.basicConfig(filename='errors.log',level=logging.ERROR)
```

```python
# Creating an custom exception class
class Invalid_age(Exception):     #Inherit the Exception class
    
    def __init__(self,msg):
        self.msg = msg
```

```python
# create a function to raise an error if age is > 150 or < 0
def Validate_age(age):
    
    if age > 140:
        raise Invalid_age("Age is much more than the normal range ")
    elif age < 0:
        raise Invalid_age("Age cannot be negative ")
    else:
        print("Age is valid")
```

```python
try:
    # taking an input from user
    age = int(input("Please Enter your Age: "))
    # using the user defined function Validate_age to check if there is any custom exception shall be raised or not.
    Validate_age(age)
except Invalid_age as e:
    logging.error(e)
```

###  Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

#### LookupError class is used to raise an error when there is an index or a key is not found for a sequence or dictionary. The demonstration for this is given below:

#### IndexError

```python
try:
    lst = [1,2,3,4,5,6]
    print(lst[9])
except LookupError as e:
    logging.error(e)
```

#### KeyError

```python
try:
    d = {'name':'Harry',
         'class' : 'Data Science',
         'id' : 32334
        }
    print(d["haello"])
except LookupError as e:
    logging.error("There is an KeyError:{}".format(e))
```

### Q5. Explain ImportError. What is ModuleNotFoundError?

####  ImportError is raised when the requested class is not available or not created, or even when there is circular dependency. 

#### ModuleNotFoundError is raised when the imported module is not available or created.The example is given below:

```python
try:
    import printworld
except ModuleNotFoundError as e:
    logging.error(e)
```

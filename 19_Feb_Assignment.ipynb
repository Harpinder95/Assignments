###  Q1. What is Flask Framework? What are the advantages of Flask Framework?


####  Flask is a lightweight Python web framework that is used for building web applications. Flask is a micro-framework, which means it is minimalistic and does not come with many built-in features. However, Flask is highly extensible, and developers can easily add third-party libraries and extensions to add more functionality.

#### Advantages :
#### 1.Simplicity: Flask is a lightweight framework that is easy to learn and use, making it an excellent choice for beginners and small projects.
#### 2. Extensibility: Flask can be easily extended using third-party libraries and extensions. Flask extensions are available for many different use cases, such as authentication, database integration, and form handling.
#### 3. Testing: Flask has built-in support for testing, making it easy to write and run tests for web applications.

###  Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

### Q3. What is App routing in Flask? Why do we use app routes?

#### App routing in Flask refers to the mechanism of associating URL paths with specific functions in a Flask web application. When a user makes a request to a specific URL path, Flask uses the associated function to generate a response. App routing is a fundamental feature of Flask and is used to define the structure and behavior of a web application.
#### We use app routes in Flask for several reasons:
#### 1. Handling HTTP requests: App routes allow us to map specific URL paths to functions in the Flask application that can handle the incoming HTTP request and generate an appropriate response.
#### 2. Enabling navigation: By defining app routes, we can create navigation links within a Flask web application. These links can be used to direct users to specific pages or functions within the application.
#### 3. Building RESTful APIs: App routes are an essential component of building RESTful APIs in Flask. Each API endpoint is associated with a specific app route, and when a user makes a request to that route, the Flask application generates a response with the requested data.

###  Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:

### Company Name: ABC Corporation

### Location: India

### Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

# Welcome route
@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

# Home route
@app.route('/')
def home():
    return 'Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999'

if __name__ == '__main__':
    app.run(host = '0.0.0.0')
```

###  Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

```python
from flask import Flask, url_for

app = Flask(__name__)

# Welcome route
@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

# Home route
@app.route('/')
# url_for()
def home():
    return 'Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999<br><a href="' + url_for('welcome') + '">Click here to visit the welcome page</a>'

if __name__ == '__main__':
    app.run(host = '0.0.0.0')
```
